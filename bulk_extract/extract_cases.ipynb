# Bulk case download and extract

Simple example of getting started with the data

```python
import os
import sys
sys.path.append('..')

import lzma
import json
import pandas as pd

from config import settings
import utils
```

```python
compressed_file = utils.get_and_extract_from_bulk(jurisdiction="Illinois", data_format="json")
```

```python
cases = []
print("File path:", compressed_file)
with lzma.open(compressed_file) as infile:
    for line in infile:
        record = json.loads(str(line, 'utf-8'))
        cases.append(record)

print("Case count: %s" % len(cases))
```

```python
df = pd.DataFrame(cases)
df.head()
```
